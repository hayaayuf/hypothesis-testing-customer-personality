## Hypothesis Testing

```python
df_update['Education'].value_counts()
```

### Step 1 : Making Hypothesis


*   H0 = Education tidak berpengaruh terhadap income = Species memiliki rata-rata
income yang sama
* H1 = Education berpengaruh terhadap income = Species memiliki rata-rata income yang berbeda signifikan

```python
df_graduation = df_update[df_update['Education'] == 'Graduation']
df_phd = df_update[df_update['Education'] == 'PhD']
df_master = df_update[df_update['Education'] == 'Master']
df_2nCycle = df_update[df_update['Education'] == '2n Cycle']
df_basic = df_update[df_update['Education'] == 'Basic']
```

### Step 2

```python
df_update.groupby(['Education'])['Income'].mean()
```

### Step 3 : Finding the P-value

```python
anova_test = st.f_oneway(df_graduation['Income'],
                         df_phd['Income'],
                         df_master['Income'],
                         df_2nCycle['Income'],
                         df_basic['Income']
                         )
```

```python
anova_test.pvalue
```

### Step 4 : Deciding

```python
if anova_test.pvalue>0.05:
    print('Species memiliki rata-rata income yang sama')
else:
    print('Species memiliki rata-rata income yang berbeda signifikan')
```
